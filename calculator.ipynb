```python
def Add():
    print("Add")

def Sub():
    print("Sub")

def Mul():
    print("Mul")

def Div():
    print("Div")

# Input
x = int(input("Enter the first number: "))
y = int(input("Enter the second number: "))


Add()
print("Addition:", x + y)


Sub()
print("Subtraction:", x - y)

Mul()
print("Multiplication:", x * y)


Div()

if y != 0:
    print("Division:", x / y)
else:
    print("Cannot divide by zero.")
```
